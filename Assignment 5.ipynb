# Cleaning Up The Data

## First I will index the data frame by Origin Airport name and remove any data that does not apply to my query 

```python
weather = weather.set_index('origin')
weather = weather[['month','day','year','wind_gust']]
weather
```

## Below I remove any rows that do not relate to the date Feb 12th 2013

```python
weather1 = weather[(weather['month']==2) & (weather['day']==12) & (weather['year']==2013)]
weather1
```

## Lastly I will sort the data frame by the largest wind gust entry

```python
windy = weather1.sort_values(by='wind_gust', ascending=False).head()
windy
```

### The chart above shows the 5 highest wind gust on Feb 12th 2013, Newark airport had the windiest weather
